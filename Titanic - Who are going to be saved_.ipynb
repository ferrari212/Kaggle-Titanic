Let's makes a function witch has as variable a binomial for each value of age, gender, and class (x_1, x_2, x_3) depending if the value is true or not, this function will be *f = ax1 + bx2 + cx3*. The *f* classifies how likely the element is to get saved. To put in proportion each of the elements, let's get sqrt(a^2+b^2+c^2) = 1. For the starting point let's get a=1, b=1, c=1.

```python
train_data['a'] = 1
train_data['b'] = 1
train_data['c'] = 1
```

```python
train_data.loc[train_data['Pclass'] == 1, 'x1'] = 1
train_data.loc[train_data['Pclass'] != 1, 'x1'] = 0
train_data.loc[train_data['Sex'] == 'female', 'x2'] = 1
train_data.loc[train_data['Sex'] != 'female', 'x2'] = 0
train_data.loc[train_data['Age'] < 18, 'x3'] = 1
train_data.loc[train_data['Age'] >= 18, 'x3'] = 0
```

```python
train_data['decision_function'] = train_data['x1']*train_data['a']+train_data['x1']*train_data['a']+train_data['x1']*train_data['a']

# for ind,row in train_data.iterrows():
#     print (row['Pclass'])
```
